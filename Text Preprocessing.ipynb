###  Can we create a function to do this?

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemmed_text(text):
    tokens = word_tokenize(text)
    stemmer = PorterStemmer()
    return [stemmer.stem(word) for word in tokens]

text2 = "The children were running quickly, they will be playing until evening."
stemmed_words = stemmed_text(text2)
print("Stemmed Words:", stemmed_words)
```

### 2. Manual Slang and Abbreviation Correction


```python
import re

# Dictionary of common slang and abbreviations
corrections = {
    "gr8": "great",
    "u": "you",
    "r": "are",
    "l8r": "later",
    "btw": "by the way",
    "omg": "oh my god",
    "cant": "can't",
    "blv": "believe"
}

# Function to apply manual corrections
def correct_text(text):
    for incorrect, correct in corrections.items():
        text = re.sub(rf'\b{incorrect}\b', correct, text, flags=re.IGNORECASE)
    return text

# Example usage
textx = "OMG!! I cant blv u got the job!!! thats gr8! I'm so happi for u!"
corrected_manual = correct_text(textx)
print("Manually Corrected Text:\n", corrected_manual)
```

### 4. Removing URLs, Mentions, and Hashtags

```python
# Text with URL, mentions, hashtags
social_text = "Check out my blog at http://example.com! Follow me on Twitter @username #excited #python"

# Remove URLs
text_no_urls = re.sub(r'http[s]?://\S+', '', social_text)

# Remove mentions (e.g., @username)
text_no_mentions = re.sub(r'@\w+', '', text_no_urls)

# Remove hashtags (e.g., #excited)
text_cleaned = re.sub(r'#\w+', '', text_no_mentions)

print("Cleaned Text (no URLs, mentions, hashtags):\n", text_cleaned)
```

### Complete Text Preprocessing Pipeline (All Steps Combined)

```python
import string
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample input text
text = "OMG!!  I cant believed u getting the job!!! thats gr8! ðŸŽ‰ I'm ðŸ˜± so happi for u! Follow @user #excited http://example.com ðŸ˜Š"

# Dictionary for correcting common slang or abbreviations
corrections = {
    "gr8": "great",
    "u": "you",
    "r": "are",
    "l8r": "later",
    "btw": "by the way",
    "omg": "oh my god",
    "cant": "can't"
}

# Step 1: Convert text to lowercase
text = text.lower()

# Step 2: Apply slang and abbreviation corrections
for incorrect, correct in corrections.items():
    text = re.sub(rf'\b{incorrect}\b', correct, text, flags=re.IGNORECASE)

# Step 3: Tokenize text
tokens = word_tokenize(text)

# Step 4: Remove punctuation
tokens = [word for word in tokens if word not in string.punctuation]

# Step 5: Remove emojis and special characters (keep alphanumerics and spaces)
tokens = [word for word in tokens if re.sub(r'[^\w\s,]', '', word) == word]

# Step 6: Remove URLs
tokens = [word for word in tokens if not re.match(r'http[s]?://', word)]

# Step 7: Remove mentions (e.g., @user)
tokens = [word for word in tokens if not word.startswith("@")]

# Step 8: Remove hashtags (e.g., #excited)
tokens = [word for word in tokens if not word.startswith("#")]

# Step 9: Remove stopwords
stop_words = set(stopwords.words("english"))
tokens = [word for word in tokens if word not in stop_words]

# Step 10: Lemmatize tokens
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(word, pos='v') for word in tokens]

# Final step: Join tokens into a cleaned string
cleaned_text = " ".join(tokens)

# Output the result
print("Cleaned Text:\n", cleaned_text)
```
